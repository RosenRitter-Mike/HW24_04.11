```python
np.random.seed(80)
rand_arr = np.random.randint(1, 101, size = (4, 4))
print('sum axis = 0 columns: \n', rand_arr.sum(axis = 0))
print('sum axis = 1 rows: \n', rand_arr.sum(axis = 1))
print('\n__rows (axis = 1)__\n')
print(f'min rows : {np.min(rand_arr, axis = 1)}')
print(f'max rows : {np.max(rand_arr, axis = 1)}')
print(f'mean rows : {np.mean(rand_arr, axis = 1)}')
print(f'median rows : {np.median(rand_arr, axis = 1)}')
print('\n__columns (axis = 0)__\n')
print(f'min columns : {np.min(rand_arr, axis = 0)}')
print(f'max columns : {np.max(rand_arr, axis = 0)}')
print(f'mean columns : {np.mean(rand_arr, axis = 0)}')
print(f'median columns : {np.median(rand_arr, axis = 0)}')
print()
print(f'nums > 5 : ', rand_arr[rand_arr > 50])
```

##### Median: 
The middle value in a sorted list of numbers. If the list has an odd count, its the central number; if even, its the average of the two middle numbers. (Example: In the list [2, 3, 5, 8, 11], the median is 5.)

##### Standard Deviation: 
A measure of how spread out numbers are from the mean (average). A low standard deviation means the values are close to the mean, while a high one means they are spread out. (Example: For the set [10, 12, 14, 16], the standard deviation is low, as values are close to the mean (13).)

##### Variance: 
The average of the squared differences from the mean, representing how much the numbers in a dataset vary. (Example: If the values are [10, 12, 14, 16], the variance would quantify the average spread from the mean.)
    

##### Variability: 
A general term for how much the data points in a set differ from each other. It encompasses measures like range, variance, and standard deviation. (Example: The list [2, 5, 10] has higher variability than [4, 5, 6].)

##### Correlation: 
A statistical measure (from -1 to +1) that describes the relationship between two variables, showing how changes in one relate to changes in the other. (Example: Height and weight have a positive correlationtaller people often weigh more.)
    

##### Covariance: 
A measure of how two variables change together. A positive covariance means they increase together, while a negative means one increases as the other decreases. (Example: Stock As price and Stock Bs price might have positive covariance if they tend to rise and fall together.)
